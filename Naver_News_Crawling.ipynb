## 전체 크롤링

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import re
import pandas as pd
import time

# 시작 날짜를 yyyymmdd 형식으로 불러오기
date_time = datetime(2015, 1, 1) # 시작 날짜 2024.12.31로 지정

# 날짜와 기사를 저장할 데이터프레임 생성
df = pd.DataFrame({'date': [], 'title': []})

# 네이버 뉴스 카테고리 설정
category1 = '101/262?date='  # 글로벌경제 카테고리
category_num = ['231', '232', '233', '234']  # 세계 카테고리 국가별 코드
category2 = '104/' + category_num[0] + '?date='  # 초기값: 첫 번째 국가 코드로 설정
category = [category1, category2]  # 두 개의 카테고리 리스트

# 2개의 대분류(글로벌경제, 세계)와 4개의 소분류(국가별)에 대해 반복
for i in range(4):
    for j in range(2):
        # 현재 소분류에 해당하는 카테고리 URL 생성
        category2 = '104/' + category_num[i] + '?date='
        main_url = 'https://news.naver.com/breakingnews/section/'
        # URL 생성
        url = main_url + category[j] + date

        # 날짜별로 기사를 수집하는 반복문
        while True:
            #현재 날짜를 yyyymmdd 형식으로 갱신
            formattedDate = date_time.strftime("%Y%m%d")
            date = formattedDate
            last_date = int(date)

            # 종료 날짜에 도달한 경우
            if last_date == 20141231:
                break

            # HTTP 요청
            html = requests.get(url)
            time.sleep(0.5)  # 서버에 부담을 줄이기 위해 0.5초 대기

            # HTTP 요청이 성공한 경우
            if html.status_code == 200:
                soup = BeautifulSoup(html.text, 'html.parser')

                # 기사 링크 수집
                tag_list = soup.find_all(class_='sa_text')  # sa_text 클래스에서 링크 추출
                sub_url_list = []
                for tag in tag_list:
                    # href 속성에서 기사 URL 추출
                    sub_url_list.append(tag.find('a')['href'])

                # 수집한 링크가 없는 경우 반복문 종료
                if not sub_url_list:
                    break

                # 리스트 길이 계산
                last_list = len(sub_url_list) - 1
                middle_list = last_list // 2  # 중간 위치 계산

                # 첫 번째, 중간, 마지막 기사 내용 추출
                try:
                    # 첫 번째 기사
                    first_html_sub = requests.get(sub_url_list[0])
                    soup_sub = BeautifulSoup(first_html_sub.text, 'html.parser')
                    first_content_title = soup_sub.find(class_='media_end_head_title').text
                    first_content_title = re.sub('[\n]', '', first_content_title)

                    # 중간 기사
                    middle_html_sub = requests.get(sub_url_list[middle_list])
                    soup_sub = BeautifulSoup(middle_html_sub.text, 'html.parser')
                    middle_content_title = soup_sub.find(class_='media_end_head_title').text
                    middle_content_title = re.sub('[\n]', '', middle_content_title)

                    # 마지막 기사
                    last_html_sub = requests.get(sub_url_list[last_list])
                    soup_sub = BeautifulSoup(last_html_sub.text, 'html.parser')
                    last_content_title = soup_sub.find(class_='media_end_head_title').text
                    last_content_title = re.sub('[\n]', '', last_content_title)

                    # 데이터프레임에 데이터 추가
                    df = pd.concat([
                        df,
                        pd.DataFrame({'date': [date, date, date],
                                      'title': [first_content_title, middle_content_title, last_content_title]})
                    ], ignore_index=True)

                except Exception as e:
                    print(f"에러 발생: {e}")
                    break  # 예외가 발생하면 반복문 종료

                # 날짜를 하루 전으로 이동
                date_time = date_time - timedelta(days=1)
```

```python
import requests
from tqdm import tqdm
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import re
import pandas as pd
import time

# 시작 날짜를 yyyymmdd 형식으로 불러오기
date_time = datetime(2015, 1, 1)  # 시작 날짜 2024.12.31로 지정

# 날짜와 기사를 저장할 데이터프레임 생성
df = pd.DataFrame({'date': [], 'title': []})

# 네이버 뉴스 카테고리 설정
category1 = '101/262?date='  # 글로벌경제 카테고리
category_num = ['231', '232', '233', '234']  # 세계 카테고리 국가별 코드

# 2개의 대분류(글로벌경제, 세계)와 4개의 소분류(국가별)에 대해 반복
total_days = (date_time - datetime(2015, 1, 1)).days
progress_bar = tqdm(total=total_days * 5, desc='크롤링 진행 중', ncols=100)

for j in range(2):  # j = 0: 글로벌경제, j = 1: 세계
    # 글로벌경제는 하나의 카테고리, 세계는 4개의 국가 코드 사용
    if j == 0:
        categories = [category1]
    else:
        categories = ['104/' + num + '?date=' for num in category_num]

    # 각 카테고리에 대해 반복
    for category in categories:
        # 날짜 초기화
        temp_date_time = date_time

        while True:
            # 마지막 날짜 도달 시 종료
            if temp_date_time.strftime('%Y%m%d') == '20141231':
                break
            formattedDate = temp_date_time.strftime("%Y%m%d")
            url = 'https://news.naver.com/breakingnews/section/' + category + formattedDate

            # HTTP 요청
            html = requests.get(url)
            time.sleep(0.5)

            # HTTP 요청이 성공한 경우
            if html.status_code == 200:
                soup = BeautifulSoup(html.text, 'html.parser')
                # sa_text 클래스에서 링크 추출
                tag_list = soup.find_all(class_='sa_text')
                sub_url_list = [tag.find('a')['href'] for tag in tag_list if tag.find('a')]

                # 링크가 없는 경우 날짜를 하루 전으로 이동
                if not sub_url_list:
                    temp_date_time -= timedelta(days=1)
                    continue

                # 리스트 길이 계산
                last_list = len(sub_url_list) - 1
                middle_list = last_list // 2

                # 첫 번째, 중간, 마지막 기사 내용 추출
                try:
                    for idx in [0, middle_list, last_list]:
                        sub_html = requests.get(sub_url_list[idx])
                        soup_sub = BeautifulSoup(sub_html.text, 'html.parser')
                        title = soup_sub.find(class_='media_end_head_title').text.strip()
                        title = re.sub('[\n]', '', title)

                        # 데이터프레임에 데이터 추가
                        df = pd.concat([
                            df,
                            pd.DataFrame({'date': [formattedDate], 'title': [title]})
                        ], ignore_index=True)
                        # 진행률 업데이트
                        progress_bar.update(1)
                except Exception as e:
                    print(f"에러 발생: {e}")

            # 날짜를 하루 전으로 이동
            temp_date_time -= timedelta(days=1)
```
